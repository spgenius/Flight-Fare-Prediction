# Exploratory Data Analysis:

From description we can see that Date_of_Journey is a object data type. Therefore, we have to convert this datatype into timestamp so as to use this column properly for prediction.

For this we require pandas to_datetime to convert object data type datetime dtype.

.dt.day method will extract only day of that date
.dt.month method will extract only month of that date



```python
train_data["Journey_Day"]= pd.to_datetime(train_data.Date_of_Journey, format="%d/%m/%Y").dt.day
```

```python
train_data["Journey_month"]=pd.to_datetime(train_data["Date_of_Journey"], format="%d/%m/%Y").dt.month
```

```python
train_data.head()
```

```python
# As we have converted Date Of Journey into Numbers, hence we can drop it.
train_data.drop(["Date_of_Journey"], axis=1, inplace=True)
```

```python
# Departure time is when a plane leaves the gate Similar to Date_of_Journey we can extract values from Dep_Time

#Extracting Hours
train_data["Dep_hour"]= pd.to_datetime(train_data["Dep_Time"]).dt.hour

#Extracting Minutes
train_data["Dep_min"]= pd.to_datetime(train_data["Dep_Time"]).dt.minute
```

```python
#Now we can drop Dep_Time as it is not required. We have converted it into numbers
train_data.drop(["Dep_Time"], axis=1, inplace=True)
```

```python
train_data.head()
```

```python
# Now we can extract hours and minutes from Arrival_time too. Similar to Dep_Time.

#Extracting hours
train_data["Arrival_hour"]= pd.to_datetime(train_data.Arrival_Time).dt.hour

#Extracting minutes
train_data["Arrival_min"]= pd.to_datetime(train_data["Arrival_Time"]).dt.minute

#Now we can drop Arrival_Time as we converted itno numeric
train_data.drop(["Arrival_Time"], axis=1, inplace=True)
```

```python
train_data.head()
```

```python
#Time Taken by a plane to reach at destination is called Duration
#It is the difference between Departure time and arrival time. Here also we need to extract hour and minuts.

# Assigned and converting duration column into list
duration = list(train_data["Duration"])
#print(duration)

for i in range(len(duration)):
    #print(duration[i].split())
    if(len(duration[i].split())!=2):   #Check if duration contains only hour and minutes
        if 'h' in duration[i]:
            duration[i]= duration[i].strip()+" 0m" #Add 0 minutes , strip() method removes any spaces or specified characters at the start and end of a string.
            #print(duration[i])
        else: 
            duration[i]= "0h "+duration[i].strip() #Add 0 hour

duration_hours=[]    #empty list to store hours
duration_mins =[]    #empty list to store mins

for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep="h")[0]))  #extract hours from duration
    duration_mins.append(int(duration[i].split(sep='m')[0].split()[-1])) #extract mins from duration
    
    #print(duration[i].split(sep='m')[0].split()[-1])
            

```

```python
train_data["duration_hours"]= duration_hours
train_data["duration_mins"]=duration_mins  # two new columns in train_data
```

```python
train_data.head()
```

```python
train_data.drop(["Duration"], axis=1, inplace=True)
```

```python
train_data.head()
```

# Handeling Categorical Data

Till now we worked to handle dates and time,now lets focus on categorical data.

One can find many ways to handle categorical data. Some of them categorical data are:

1. Nominal data --> data are not in order --> OneHotEncoder is used in this case
2. Ordinal Data --> data is in order --> LabelEncoder is used in this case

```python
# Airline column, it is Nominal Data as there is no order
train_data["Airline"].value_counts()
```

```python
# Airline Vs Price

sns.catplot(y='Price', x='Airline', data=train_data.sort_values("Price",ascending=False),kind='boxen', height=6, aspect=3)

#From graph we can see that Jet Airways Business have the highest Price. So this is kind of outlier. 
#Apart from the first Airline almost all are having similar median
```

```python
# As Airline is Nominal Categorical Data we will perform OneHotEncoding

Airline= train_data[["Airline"]]
Airline=pd.get_dummies(Airline, drop_first=True)
Airline.head()
```

```python
# Now lets check Source column
train_data["Source"].value_counts()
```

```python
#Source Vs Price

sns.catplot(y='Price', x='Source', data=train_data.sort_values("Price", ascending=False), kind='boxen',height=6, aspect=3)
```

```python
# As source is Nominal Category, hence use OneHotEncoding
Source= train_data[["Source"]]
Source= pd.get_dummies(Source, drop_first=True)
Source.head()
```

```python
# Destination Folder
train_data["Destination"].value_counts()
```

```python
# Destination also the Nominal Category, use OneHotEncoding
Destination= train_data[["Destination"]]
Destination=pd.get_dummies(Destination, drop_first=True)
Destination.head()
```

```python
# Now lets see route column
train_data["Route"]
```

```python
# Addtional_Info column
#sns.countplot(x="Additional_Info", data=train_data )
train_data["Additional_Info"].value_counts()
```

```python
# Additional_Info contains almost 80% no_info
# And Route & Total_Stops are related to each other, hence we can drop Route and Additional_Info
train_data.drop(["Route", "Additional_Info"], axis=1, inplace=True)
```

```python
train_data.head()
```

```python
#Total_Stops column
train_data["Total_Stops"].value_counts()
```

```python
# As this is the case of Ordinal Categorical type we perform LabelEncoder
# Here values are assigned with corresponding keys
train_data.replace({"non-stop":0, "1 stop":1, "2 stops":2,"3 stops":3, "4 stops":4}, inplace=True) #These values are replaced in place
```

```python
train_data.head()
```

```python
# Now lets combine all the features train_data set which we got during encoding
final_data=pd.concat([train_data, Airline, Source, Destination], axis=1)
```

```python
#Lets drop categorical columns which we have taken care of conversion
final_data.drop(["Airline","Source","Destination"], axis=1, inplace=True)
```

# Now Lets do Same preprocessing for Test Set

Why do we need to do same process for Test Set separatly, why we can not combine Test set and Train Set and do preprocessing at once?

The reason is data leakage.


```python
test_data= pd.read_excel("Test_set.xlsx")
test_data.head()
```

```python
#Preprocessing

print("Test data Info:")
print("-"*75)
print(test_data.info())

print("\nNull Values:")
print("-"*75)
print(test_data.isnull().sum())

#No null values, lets move to EDA

# Date_of_Journey
test_data["Journey_day"] = pd.to_datetime(test_data.Date_of_Journey, format="%d/%m/%Y").dt.day
test_data["Journey_month"] = pd.to_datetime(test_data["Date_of_Journey"], format = "%d/%m/%Y").dt.month
test_data.drop(["Date_of_Journey"], axis = 1, inplace = True)

# Dep_Time
test_data["Dep_hour"] = pd.to_datetime(test_data["Dep_Time"]).dt.hour
test_data["Dep_min"] = pd.to_datetime(test_data["Dep_Time"]).dt.minute
test_data.drop(["Dep_Time"], axis = 1, inplace = True)

# Arrival_Time
test_data["Arrival_hour"] = pd.to_datetime(test_data.Arrival_Time).dt.hour
test_data["Arrival_min"] = pd.to_datetime(test_data.Arrival_Time).dt.minute
test_data.drop(["Arrival_Time"], axis = 1, inplace = True)

# Duration
duration = list(test_data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration

# Adding Duration column to test set
test_data["Duration_hours"] = duration_hours
test_data["Duration_mins"] = duration_mins
test_data.drop(["Duration"], axis = 1, inplace = True)

# Categorical data

print("\nAirline")
print("-"*75)
print(test_data["Airline"].value_counts())
Airline = pd.get_dummies(test_data["Airline"], drop_first= True)

print()

print("\nSource")
print("-"*75)
print(test_data["Source"].value_counts())
Source = pd.get_dummies(test_data["Source"], drop_first= True)

print()

print("\nDestination")
print("-"*75)
print(test_data["Destination"].value_counts())
Destination = pd.get_dummies(test_data["Destination"], drop_first = True)

# Additional_Info contains almost 80% no_info
# Route and Total_Stops are related to each other
test_data.drop(["Route", "Additional_Info"], axis = 1, inplace = True)

# Replacing Total_Stops
test_data.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)

# Concatenate dataframe --> test_data + Airline + Source + Destination
data_test = pd.concat([test_data, Airline, Source, Destination], axis = 1)

data_test.drop(["Airline", "Source", "Destination"], axis = 1, inplace = True)


print("\nShape of test data : ", data_test.shape)
```

## Fitting model using Random Forest

1. Split dataset into train and test set in order to prediction with respect to X_test
2. If needed do scaling of data
    * Scaling is not done in Random forest
3. Import model
4. Fit the data
5. Predict with respect to X_test
6. In regression check **RSME** Score
7. Plot graph

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# It is s regression problem, hence will apply RandomForestRegressor model

from sklearn.ensemble import RandomForestRegressor
reg_rf = RandomForestRegressor()
reg_rf.fit(X_train, y_train)
```

```python
y_pred = reg_rf.predict(X_test)
```

```python
#Training data score
reg_rf.score(X_train, y_train)
```

```python
#Testing data score
reg_rf.score(X_test, y_test)
```

```python
sns.distplot(y_test-y_pred)
plt.show()

# As our Gaussian Dstribution is pretty good means our model is good.
```

```python

plt.scatter(y_test, y_pred, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

```python
#Lets check the Mean Square Error, Mean Absolute Error and Root Mean Square Error
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

## Hyperparameter Tuning


* Choose following method for hyperparameter tuning
    1. **RandomizedSearchCV** --> Fast
    2. **GridSearchCV**
* Assign hyperparameters in form of dictionery
* Fit the model
* Check best paramters and best score

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = reg_rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
#check the best parameters
rf_random.best_params_
```

```python
prediction = rf_random.predict(X_test)
```

```python
#lets check again outliers
plt.figure(figsize = (8,8))
sns.distplot(y_test-prediction)
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.scatter(y_test, prediction, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```
